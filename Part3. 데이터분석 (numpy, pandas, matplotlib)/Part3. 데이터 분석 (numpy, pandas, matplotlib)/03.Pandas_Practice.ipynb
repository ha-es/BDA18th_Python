```python
# 1. 'pH' 값이 가장 높은 행을 출력하세요.
max_pH_row = wine_df.loc[wine_df['pH'].idxmax()]
print(max_pH_row)

# 2. 'alcohol' 값이 7.5 이상인 행들의 'sulphates' 평균값을 계산하세요.
mean_sulphates = wine_df.loc[wine_df['alcohol'] >= 7.5, 'sulphates'].mean()
print(mean_sulphates)

# 3. 'quality' 값이 7 이상인 행들의 'alcohol' 평균값을 계산하세요.
mean_alcohol_high_quality = wine_df.loc[wine_df['quality'] >= 7, 'alcohol'].mean()
print(mean_alcohol_high_quality)

# 5. 'residual_sugar' 값이 1 이하인 행들의 개수를 계산하세요.
count_low_residual_sugar = len(wine_df.loc[wine_df['residual sugar'] <= 1])
print(count_low_residual_sugar)

# 4. 'volatile_acidity' 값이 가장 높은 3개의 행을 출력하세요.
top_3_volatile_acidity = wine_df.nlargest(3, 'volatile acidity')
print(top_3_volatile_acidity)
```

```python
# 연습문제 8: 데이터프레임에서 각 와인의 알콜(alcohol)별로 평균 당도(residual sugar)를 구해보세요.
# color별 평균 alcohol과 residual sugar 구하기
mean_by_color = wine_df.groupby('alcohol')['alcohol', 'residual sugar'].mean()

# 결과 출력
print(mean_by_color)
```

```python
# 연습문제 10: 데이터프레임에서 각 와인의 색깔(color)과 알코올 도수(alcohol)에 따라서 최대 산도(fixed acidity)를 구해보세요.
# 'fixed acidity' 열의 값이 7 이하인 데이터 필터링
low_acidity_data = wine_df[wine_df['fixed acidity'] <= 7]

# 'quality' 열 값이 0인 데이터로 변경
low_acidity_data.loc[:, 'quality'] = 0

# 'fixed acidity' 열의 값이 8 이상인 데이터 필터링
high_acidity_data = wine_df[wine_df['fixed acidity'] >= 8]

# 'quality' 열 값이 2인 데이터로 변경
high_acidity_data.loc[:, 'quality'] = 2

# 변경된 데이터를 원래 데이터프레임에 합치기
changed_data = pd.concat([low_acidity_data, high_acidity_data], ignore_index=True)

print("fixed acidity 열의 값이 7 이하인 데이터의 quality 열 값이 0으로, 8 이상인 데이터의 quality 열 값이 2로 변경 완료")
```

## 3. 서울시 인구 현황 분석

```python
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/data/5.pandas/seoul_popular_data.xls'

df_seoul = pd.read_excel(file_path)
df_seoul
```

```python
df_seoul = pd.read_excel(file_path, header=1, usecols="B, D, G, J, N")
df_seoul
```

```python
# 컬럼명 변경
# (자치구 -> 구, 합계 -> 인구수, 등록외국인 -> 외국인, 65세이상고령자->고령자)

df_seoul.columns = ["구", "인구수", "한국인","외국인", "고령자"]
df_seoul
```

```python
# 합계행 제거
df_seoul.drop([0], inplace=True)
```

```python
df_seoul.drop([1], inplace=True)
```

```python
df_seoul
```

```python
df_seoul.info()
```

```python
df_seoul['인구수'] = df_seoul['인구수'].astype('int')
df_seoul['한국인'] = df_seoul['한국인'].astype('int')
df_seoul['외국인'] = df_seoul['외국인'].astype('int')
df_seoul['고령자'] = df_seoul['고령자'].astype('int')
```

```python
df_seoul.info()
```

```python
# (1) 각 구별 전체 인구를 이용해서 구별 '외국인비율'과 '고령자비율' 컬럼을 추가하고,
# (2) 소수점 첫째자리 반올림하기
### - 외국인 비율: 외국인 / 인구수 * 100
### - 고령자 비율: 고령자 / 인구수 * 100

df_seoul['외국인비율'] = round(df_seoul['외국인'] / df_seoul['인구수'] * 100, 1)
df_seoul['고령자비율'] = round(df_seoul['고령자'] / df_seoul['인구수'] * 100, 1)
df_seoul
```

```python
# (5) 인구수별로 내림차순 정렬하기
df_seoul.sort_values(by='인구수', ascending=False)
```

```python
# (6) 외국인 / 외국인비율로 내림차순 정렬하기
df_seoul.sort_values(by='외국인', ascending=False)
```

```python
df_seoul.sort_values(by='외국인비율', ascending=False)
```

```python
# (7) 고령자수 / 고령자비율로 내림차순 정렬하기
df_seoul.sort_values(by='고령자')
```

```python
df_seoul.sort_values(by='고령자비율')
```
