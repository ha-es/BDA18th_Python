```python
# import
import numpy as np
```

```python
array1 = np.array([1,2,3]) # 1차원 배열 - 하나의 컬럼 데이터
array2 = np.array([[1,2,3], [1,2,3]]) # 2차원 배열 - 행과 열이 포함된 데이터

array1
array2

array1.shape
array2.shape # (2,3) => 2행 3열

array1.ndim
array2.ndim
```

```python
array3 = np.array(
    [ [ ["A", "B", "C"] ] ]
)

print(array3.ndim) # 차원
print(array3.shape) # 차원의 요소 개수 (행, 열)
print(array3.dtype) # 타입 -> 공통적인 데이터 타입
```

### ndarray 데이터 타입

```python
list2 = [1 , 2 , 'c']

arr2 = np.array(list2)
print(arr2.dtype)

list3 = [1.0, 2.0, 3]
arr3 = np.array(list3)
print(arr3.dtype)
```

```python
array4 = np.array(["1","2","3","4","5"])
array4.dtype # 문자열 데이터

array4 = array4.astype(np.int64) # astype
array4.dtype # 정수형 데이터로 변경
```

```python
array_int = np.array([1,2,3])
array_float = array_int.astype('float64') # np.float64

array_float.dtype

array_int1 = array_float.astype('int64')
array_int1.dtype

array_float1 = np.array([1.0, 2.0, 3.1])

array_int2 = array_float1.astype('int64')
print(array_int2, array_int2.dtype)
```

- Boolean Indexing

```python
arr = np.arange(10)

# print(arr)
# print(arr.dtype) # 숫자형 데이터 타입
# print(arr.shape) # (10) 10행인 1차원 배열

arr[0:5]

print(arr > 6)
print(arr[arr > 6])
```

### 인덱싱, 슬라이싱

```python
arr = np.arange(1,10)
arr

arr[0]
arr[0:5]
arr[:]
```

### sort()
- 정렬을 위해 사용하는 함수

```python
arr = np.array([10, 30, 2, 9, 6])

arr_sorted = np.sort(arr)

print(arr_sorted)
print(arr.sort())

arr_sorted_desc = np.sort(arr)[::-1] # 내림차순 [::-1]
print(arr_sorted_desc)
```

```python
arr = np.array(["Korea", "Singapole", "China", "Japan"])

arr_sorted = np.sort(arr)

print(arr_sorted)

#  유니코드 코드 포인트 값인데 맨 앞글자만 비교.
```
