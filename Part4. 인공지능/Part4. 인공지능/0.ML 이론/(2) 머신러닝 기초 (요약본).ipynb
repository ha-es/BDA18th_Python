```python
# sklearn import - 인공지능 사용에 필요한 모듈 로드
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
```

```python
iris = load_iris()

iris_data = iris.data # feature data
iris_label = iris.target # target data

df_iris = pd.DataFrame(data=iris_data, columns=iris.feature_names)
df_iris['label'] = iris_label

df_iris
```

```python
# 학습 데이터 셋과 테스트 데이터 셋이 필요합니다.
# train_test_split()

# X_train: 학습용 피처 데이터, X_test: 테스트용 피처 데이터
# y_train: 학습용 타겟 데이터, y_test: 테스트용 타겟 데이터
X_train, X_test, y_train, y_test =
train_test_split(iris_data, iris_label,
                 test_size=0.2, random_state=11)
```

```python
# print(X_train, X_test, y_train, y_test)
print(type(X_train), type(X_test), type(y_train), type(y_test))
```

```python
# 학습 데이터
print(len(X_train))
print(len(y_train))

# 테스트용 데이터
print(len(X_test))
print(len(y_test))
```

```python
# Decision Tree Classifier 알고리즘으로 학습을 시키겠다.
dt_clf = DecisionTreeClassifier(random_state=11)

# 인공지능한테 학습해 라고 명령하는 함수 => fit() # 모의고사
dt_clf.fit(X_train, y_train) # 학습용 데이터 # fit은 답을 알려준다.
```

```python
# 예측 => predict() # 수능
pred = dt_clf.predict(X_test) # 테스트 데이터 셋을 주고 에측 해보라고 시킨다. # 답을 알려주지 않는다.
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred) # 예측 정확도
```

### 교차 검증

### 일반적인 K-Fold 검증

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

iris = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)

train_data = iris.data # 피처 데이터
train_target = iris.target # 타겟 데이터
```

```python
import numpy as np

# 5개의 폴드 세트로 구분하겠다.
kfold = KFold(n_splits=5)
cv_accuracy = []

num = 0
for train_index, test_index in kfold.split(train_data):
  # 학습용 데이터 셋에서 => 학습용 데이터와 검증용 데이터를 분리하는 작업
  X_train, X_test = train_data[train_index], train_data[test_index] # 피처값
  y_train, y_test = train_target[train_index], train_target[test_index] # 타겟값

  # 학습 및 예측
  dt_clf.fit(X_train, y_train) # 학습용 데이터로 학습 진행
  pred = dt_clf.predict(X_test)

  # 정확도 체크
  accuracy = accuracy_score(y_test, pred)
  num += 1 # 횟수 체크용

  cv_accuracy.append(accuracy)

  print("{} 번째 트레이닝".format(num))
  print("검증 정확도 : ", accuracy, "학습 데이터 크기 : ", X_train.shape[0], "검증 데이터 크기: ", X_test.shape[0])
  print("검증 데이터 셋 인덱스 값 : ", test_index)

  print("===" * 10)


print("최종 정확도 평균 값 : ", np.mean(cv_accuracy))
```

```python
import pandas as pd

iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df['label'].value_counts()

kfold = KFold(n_splits=3)

# kfold.split(X)는 폴드 세트를 3번 반복할 때마다 달라지는 학습/테스트 용 데이터 로우 인덱스 번호 반환.
n_iter =0
for train_index, test_index  in kfold.split(iris_df):
    n_iter += 1
    label_train= iris_df['label'].iloc[train_index]
    label_test= iris_df['label'].iloc[test_index]
    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3)

num = 0
for train_index, test_index in skf.split(df_iris, df_iris['label']):
  label_train = df_iris['label'].iloc[train_index]
  label_test = df_iris['label'].iloc[test_index]
  num += 1

  print("{} 번째 트레이닝".format(num))
  print("학습 레이블 인덱스 값 : \n", label_train.value_counts())
  print("검증 레이블 인덱스 값 : \n", label_test.value_counts())
  print("===" * 10)
```

### cross_val_score() : 교차 검증을 쉽게 도와주는 함수

```python
from sklearn.model_selection import cross_val_score

iris = load_iris()
dt_clf = DecisionTreeClassifier(random_state=121)

features = iris.data # x
target = iris.target # y

scores = cross_val_score(dt_clf, features, target, scoring="accuracy", cv=3)
scores
print("각 검증별 정확도 데이터 : ", scores)
print("평균 정확도 데이터 : ", np.mean(scores))
```
