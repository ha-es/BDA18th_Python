# 4. 딕셔너리형 (Dict)

Dict → 사전

- { 단어 : 뜻 }의 형태

중괄호 안에 키(key)와 벨류(value)로 구성되어 있는 형태

데이터의 빠른 검색, 추가, 삭제를 지원

{Key1:Value1, Key2:Value2, Key3:Value3, ...}

x = {"name" : "tony", "age" : "25", "country" : "korea"}
print(x["name"])

value에는 리스트 형태도 가능
x = {"name" : "tony", "age" : "25", "country" : ["korea", "usa"]}

단, 키는 중복이 안된다.

- dict => API, json
- 컴퓨터 간의 의사소통 방식 (프로토콜)

## (1) 딕셔너리의 주요 특징

1. **키-값 쌍으로 구성**:
    - 데이터는 고유한 키와 값의 쌍으로 저장됩니다.
    - 예: `{"name": "Alice", "age": 25}`
2. **키는 고유**:
    - 키는 중복될 수 없습니다.
3. **변경 가능 (Mutable)**:
    - 딕셔너리는 생성 후 수정, 추가, 삭제가 가능합니다.
4. **키는 해시 가능한 값만 사용 가능**:
    - 숫자, 문자열, 튜플 등 불변 자료형이 키로 사용됩니다.

## (2) 딕셔너리 생성 및 기본 사용법

### **딕셔너리 생성**

```python
# 빈 딕셔너리 생성
empty_dict = {}

# 값이 있는 딕셔너리 생성
person = {"name": "Alice", "age": 25, "city": "New York"}

print(person)  # {'name': 'Alice', 'age': 25, 'city': 'New York'}

print(type(person))
```

### **딕셔너리 키로 값 접근**

```python
person = {"name": "Alice", "age": 25, "city": "New York"}

# 키로 값 가져오기
print(person["name"])  # Alice
print(person["age"])   # 25
```

### **딕셔너리 값 추가 및 수정**

```python
person = {"name": "Alice", "age": 25}

# 값 추가
person["city"] = "New York"
print(person)  # {'name': 'Alice', 'age': 25, 'city': 'New York'}

# 값 수정
person["age"] = 26
print(person)  # {'name': 'Alice', 'age': 26, 'city': 'New York'}
```

### **딕셔너리 값 삭제**

```python
person = {"name": "Alice", "age": 25, "city": "New York"}

# 특정 키 삭제
del person["city"]
print(person)  # {'name': 'Alice', 'age': 25}

# 모든 키-값 삭제
person.clear()
print(person)  # {}
```

## (3) 딕셔너리의 주요 메서드

### **1. `keys()`, `values()`, `items()`**

```python
person = {"name": "Alice", "age": 25, "city": "New York"}

print(person.keys())   # dict_keys(['name', 'age', 'city'])
print(person.values()) # dict_values(['Alice', 25, 'New York'])
print(person.items())  # dict_items([('name', 'Alice'), ('age', 25), ('city', 'New York')])
```

### **2. `get()`**

```python
person = {"name": "Alice", "age": 25}

print(person.get("name"))          # Alice
print(person.get("city", "Unknown"))  # Unknown
```

### **3. `update()`**

```python
person = {"name": "Alice", "age": 25}
person.update({"city": "New York", "age": 26})

print(person)  # {'name': 'Alice', 'age': 26, 'city': 'New York'}
```

### **4. `pop()`**

```python
person = {"name": "Alice", "age": 25}
age = person.pop("age")

print(age)     # 25
print(person)  # {'name': 'Alice'}
```
