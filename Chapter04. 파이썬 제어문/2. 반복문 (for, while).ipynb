
## for문의 기본 형태

```python
for 변수 in 리스트(또는 튜플, 문자열):
    실행할 문장1
```



## (1) `for` 반복문

`for` 반복문은 시퀀스(리스트, 문자열, 튜플 등)의 각 요소를 순차적으로 가져와 작업을 수행합니다.

```python
for 변수 in 시퀀스:
    실행할 코드
```


### 리스트

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

### range()

````python
for i in range(5): # 숫자(0부터 5미만)
    print(i)
````

### 문자열

````python
text = "Python"
for char in text:
    print(char)
````

## (2) `while` 반복문

- `while` 문은 조건이 참(True)인 동안 코드를 반복 실행합니다.

while 조건:
    실행할 코드

````python
count = 0
while count < 3:
    print("반복:", count)

    count += 1 # count = count + 1
````

조건을 만족하지 않으면 실행되지 않음

````python
count = 10
while count < 3:
    print("이 코드는 실행되지 않습니다.")
````


## (3) 반복문에서 자주 사용되는 키워드

### 1. break
- `break`는 반복문을 **즉시 종료**합니다.


````python
for i in range(5):
    if i == 3:
        break
    print(i)
````

### 2. continue
- continue는 현재 반복을 건너뛰고, 다음 반복으로 넘어갑니다.

````python
for i in range(5):
    if i == 3:
        continue
    print(i)
````

### 3. pass
- pass는 아무 작업도 하지 않고 반복문을 유지합니다.

````python
for i in range(5):
    if i == 3:
        pass # todo: 여기서부터
    print(i)
````


## (4) 반복문에서 자주 사용되는 함수

### 1. `range(start, stop, step)`
- 숫자 범위를 생성하며, 시작값(start), 끝값(stop), 증가값(step)을 지정할 수 있습니다.


````python
for i in range(1, 9, 2):
    print(i)
````

### 2. `enumerate()`
- 리스트나 문자열을 순회하며 인덱스와 값을 동시에 가져옵니다.

````python
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(index, fruit)
````

### 3. `zip()`
- 여러 시퀀스를 동시에 순회합니다.

````python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f"{name} is {age} years old.")
````

### 4. `reversed()`
- 리스트나 문자열을 역순으로 순회합니다.

````python
for char in reversed([1,2,3,4,5]):
    print(char)
````
