# 1. 조건문 (if)

특정 조건을 검사하여 그 조건에 따라 다른 코드를 실행하도록 하는 구문

**`if`, `elif`, `else`** 키워드를 사용하여 조건문을 작성

프로그램의 흐름을 제어

````python
# 조건문 예제
x = 5

if x > 5:
    print("x는 5보다 큽니다.")  # 출력됨
elif x == 5: # else if => elif
    print("x는 5입니다.")
else:
    print("x는 5보다 작습니다.")
````

## (1) 조건문의 구조

### **1. 기본 구조**

```python
if 조건:
    실행할 코드
elif 조건:
    실행할 코드
else:
    실행할 코드
```

### **2. 작동 원리**

- **`if`**: 조건이 참이면 해당 블록 실행.
- **`elif`**: 위 조건이 모두 거짓일 때 새로운 조건을 검사.
- **`else`**: 모든 조건이 거짓일 때 실행.

## (2) 조건문 예제

### **1. 기본 사용법**

````python
x = 10

if x > 5:
    print("x는 5보다 큽니다.")  # 출력됨
elif x == 5:
    print("x는 5입니다.")
else:
    print("x는 5보다 작습니다.")
````

### **2. 중첩 조건문**

````python
x = 15

if x > 10:
    if x % 2 == 0:
        print("x는 10보다 크고 짝수입니다.")
    else:
        print("x는 10보다 크고 홀수입니다.")
else:
    print("x는 10 이하입니다.")
````

### **3. 조건문**

````python
age = 25
is_student = True

if age < 18 or is_student:
    print("할인이 적용됩니다.")  # 출력됨
else:
    print("할인이 적용되지 않습니다.")
````

## (3) 조건문과 자주 사용되는 함수

### **1. `input()`**

사용자로부터 입력을 받아 조건을 검사할 때 자주 사용됩니다.

````python
age = int(input("나이를 입력하세요: "))

if age >= 18:
    print("성인입니다.")
else:
    print("미성년자입니다.")
````

### **2. `len()`**

문자열이나 리스트의 길이를 조건으로 사용할 수 있습니다.

````python
password = "abc123"

if len(password) >= 6:
    print("비밀번호가 유효합니다.")
else:
    print("비밀번호는 최소 6자 이상이어야 합니다.")
````

## (4) 비교연산자

조건문에 값을 비교할 때 사용

- a와 b가 같다 : `a == b`
- a와 b가 같지 않다 : `a != b`
- a가 b보다 작다 : `a < b`
- a가 b보다 작거나 같다 : `a <= b`
- a가 b보다 크다 : `a > b`
- a가 b보다 크거나 같다 : `a >= b`


````python
a = 1
b = 3

print(a == b)  # False
print(a != b)  # True
print(a < b)   # True
print(a <= b)  # True
print(a > b)   # False
print(a >= b)  # False
````

## **(5) and, or 연산자**

airplane = False
