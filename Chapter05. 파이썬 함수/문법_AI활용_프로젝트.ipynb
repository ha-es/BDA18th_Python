```python
import random

def generate_question(question_type, level):
    """난이도와 문제 유형에 따라 문제를 생성하는 함수"""
    if question_type == "어휘":
        words = {"easy": ["apple", "book", "table"],
                 "medium": ["persuade", "achieve", "construct"],
                 "hard": ["meticulous", "exacerbate", "quintessential"]}
        word = random.choice(words[level])
        return f"다음 단어의 뜻을 적으세요: {word}", word
    
    elif question_type == "문법":
        grammar = {"easy": [("He ___ a student.", "is"),
                              ("She ___ to school every day.", "goes")],
                   "medium": [("If I ___ you, I would study harder.", "were"),
                               ("He said he ___ there before.", "had been")],
                   "hard": [("Not only ___ he smart, but he is also kind.", "is"),
                            ("Had I known the truth, I ___ acted differently.", "would have")]}
        question, answer = random.choice(grammar[level])
        return question, answer
    
    elif question_type == "독해":
        passages = {"easy": ["The cat is on the mat. What is on the mat?", "cat"],
                    "medium": ["John likes apples. He eats one every day. What does John eat?", "apple"],
                    "hard": ["Despite the inclement weather, the event proceeded as planned. Did the event continue?", "yes"]}
        passage, answer = random.choice(passages[level])
        return passage, answer
    
    return "잘못된 선택입니다.", ""

def quiz():
    """퀴즈 실행 함수"""
    score = 0
    while True:
        print("\n[ 문제 유형 선택 ]")
        print("1. 어휘")
        print("2. 문법")
        print("3. 독해")
        print("4. 종료")
        
        choice = input("번호를 입력하세요: ")
        if choice == "4":
            print(f"최종 점수: {score}")
            break
        
        types = {"1": "어휘", "2": "문법", "3": "독해"}
        question_type = types.get(choice, None)
        if not question_type:
            print("잘못된 입력입니다.")
            continue
        
        print("\n[ 난이도 선택 ]")
        print("1. 쉬움")
        print("2. 보통")
        print("3. 어려움")
        
        level_choice = input("번호를 입력하세요: ")
        levels = {"1": "easy", "2": "medium", "3": "hard"}
        level = levels.get(level_choice, None)
        if not level:
            print("잘못된 입력입니다.")
            continue
        
        question, answer = generate_question(question_type, level)
        print("\n문제:", question)
        user_answer = input("답을 입력하세요: ")
        
        if user_answer.lower() == answer.lower():
            print("정답입니다! +1점")
            score += 1
        else:
            print(f"오답입니다. 정답: {answer}")

if __name__ == "__main__":
    quiz()
```

```python
import tkinter as tk
from tkinter import messagebox
import random

class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("문제은행 자동 생성 프로그램")
        self.score = 0
        
        self.label = tk.Label(root, text="[ 문제 유형 선택 ]", font=("Arial", 14))
        self.label.pack(pady=10)
        
        self.btn_vocab = tk.Button(root, text="어휘", command=lambda: self.select_type("어휘"), width=20, height=2)
        self.btn_vocab.pack()
        self.btn_grammar = tk.Button(root, text="문법", command=lambda: self.select_type("문법"), width=20, height=2)
        self.btn_grammar.pack()
        self.btn_reading = tk.Button(root, text="독해", command=lambda: self.select_type("독해"), width=20, height=2)
        self.btn_reading.pack()
        self.btn_exit = tk.Button(root, text="종료", command=root.quit, width=20, height=2, fg="red")
        self.btn_exit.pack()
        
    def select_type(self, question_type):
        self.question_type = question_type
        self.label.config(text="[ 난이도 선택 ]")
        
        self.btn_vocab.pack_forget()
        self.btn_grammar.pack_forget()
        self.btn_reading.pack_forget()
        self.btn_exit.pack_forget()
        
        self.btn_easy = tk.Button(self.root, text="쉬움", command=lambda: self.generate_question("easy"), width=20, height=2)
        self.btn_easy.pack()
        self.btn_medium = tk.Button(self.root, text="보통", command=lambda: self.generate_question("medium"), width=20, height=2)
        self.btn_medium.pack()
        self.btn_hard = tk.Button(self.root, text="어려움", command=lambda: self.generate_question("hard"), width=20, height=2)
        self.btn_hard.pack()
    
    def generate_question(self, level):
        questions = {
            "어휘": {"easy": ["apple", "book", "table"],
                     "medium": ["persuade", "achieve", "construct"],
                     "hard": ["meticulous", "exacerbate", "quintessential"]},
            "문법": {"easy": [("He ___ a student.", "is"),
                              ("She ___ to school every day.", "goes")],
                     "medium": [("If I ___ you, I would study harder.", "were"),
                                 ("He said he ___ there before.", "had been")],
                     "hard": [("Not only ___ he smart, but he is also kind.", "is"),
                              ("Had I known the truth, I ___ acted differently.", "would have")]},
            "독해": {"easy": ["The cat is on the mat. What is on the mat?", "cat"],
                    "medium": ["John likes apples. He eats one every day. What does John eat?", "apple"],
                    "hard": ["Despite the inclement weather, the event proceeded as planned. Did the event continue?", "yes"]}
        }
        
        question_data = questions[self.question_type][level]
        if isinstance(question_data[0], tuple):
            question, answer = random.choice(question_data)
        else:
            question = f"다음 단어의 뜻을 적으세요: {random.choice(question_data)}"
            answer = question_data[-1]
        
        self.ask_question(question, answer)
    
    def ask_question(self, question, answer):
        self.label.config(text=question)
        
        self.entry = tk.Entry(self.root, width=30)
        self.entry.pack(pady=10)
        
        self.btn_submit = tk.Button(self.root, text="제출", command=lambda: self.check_answer(answer), width=20, height=2)
        self.btn_submit.pack()
    
    def check_answer(self, correct_answer):
        user_answer = self.entry.get().strip().lower()
        
        if user_answer == correct_answer.lower():
            self.score += 1
            messagebox.showinfo("정답!", f"정답입니다! 현재 점수: {self.score}")
        else:
            messagebox.showinfo("오답", f"오답입니다. 정답: {correct_answer}")
        
        self.entry.pack_forget()
        self.btn_submit.pack_forget()
        self.reset_quiz()
    
    def reset_quiz(self):
        self.label.config(text="[ 문제 유형 선택 ]")
        
        self.btn_easy.pack_forget()
        self.btn_medium.pack_forget()
        self.btn_hard.pack_forget()
        
        self.btn_vocab.pack()
        self.btn_grammar.pack()
        self.btn_reading.pack()
        self.btn_exit.pack()
        
if __name__ == "__main__":
    root = tk.Tk()
    app = QuizApp(root)
    root.mainloop()
```
